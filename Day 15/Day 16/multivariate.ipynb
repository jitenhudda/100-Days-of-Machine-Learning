# **Scatter Plot**  - (numerical-numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)  #bivariate
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)   #multi-variate
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', data=tips)
```

# **Bar Plot** - (numerical - categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic)
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic)
```

```python
sns.barplot(x='Pclass', y='Fare', hue='Sex', data=titanic)
```

# **Box Plot** - (numerical - categorical)

```python
sns.boxplot(x='Sex', y='Age',hue='Survived', data=titanic)
```

# **DisPlot** - (numerical - categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

# **HeatMap** - (categorical - categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

# **ClusterMap** - (categorical - categorical)

```python
sns.clustermap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```
