# Combine Everything

```python
book = soup.find("article", class_="product_pod")

title = book.h3.a["title"]
price = book.find("p", class_="price_color").text
rating = book.find("p")["class"][1]
availability = book.find("p", class_="instock availability").text.strip()
image_url = "https://books.toscrape.com/" + book.find("img")["src"].replace("../", "")

print("Title :", title)
print("Price :", price)
print("Rating :", rating)
print("Availability :", availability)
print("Image :", image_url)
```

# Find All Books

```python
books = soup.find_all("article", class_="product_pod")
print(len(books))
```

# Extract Data from All 20 Books

```python
for i in books:
    title = i.h3.a["title"]
    price = i.find("p", class_="price_color").text
    rating = i.find("p")["class"][1]

    print(title)
    print(price)
    print(rating)
    print("-" * 50)
```

# Store Data in a List

```python
data = []

for book in books:

    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    rating = book.find("p")["class"][1]
    availability = book.find("p", class_="instock availability").text.strip()
    image_url = "https://books.toscrape.com/" + book.find("img")["src"].replace("../", "")


    data.append({
        "Title": title,
        "Price": price,
        "Rating": rating,
        "Availability": availability,
        "Image URL": image_url,
    })
```

```python
print(data)
```

# Create a DataFrame

```python
df = pd.DataFrame(data)
df
```

# Save as CSV

```python
df.to_csv("books_page_1.csv", index=False)
print("CSV Saved Successfully!")
```

# Scrap all pages

```python
df = pd.DataFrame(columns=[
    "Title",
    "Price",
    "Rating",
    "Availability",
    "Image URL"
])

for page in range(1, 51):

    url = f"https://books.toscrape.com/catalogue/page-{page}.html"

    response = requests.get(url)

    soup = BeautifulSoup(response.text, "lxml")

    books = soup.find_all("article", class_="product_pod")

    for book in books:

        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text
        rating = book.find("p")["class"][1]
        availability = book.find("p", class_="instock availability").text.strip()
        image_url = "https://books.toscrape.com/" + book.find("img")["src"].replace("../", "")

        df.loc[len(df)] = [
            title,
            price,
            rating,
            availability,
            image_url
        ]

    print(f"Page {page} Completed")

print("Scraping Finished!")
```
